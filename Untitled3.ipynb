```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

IMAGE_NAME = "ornek4.jpeg"

def create_references():

    if not os.path.exists(IMAGE_NAME):
        print(f"ERROR: The file '{IMAGE_NAME}' was not found! Please upload it using the file menu on the left.")
        return

    img = cv2.imread(IMAGE_NAME)
    if img is None:
        print("Error: Image could not be read.")
        return

    # Preprocessing
    imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    imgBlur = cv2.GaussianBlur(imgGray, (5, 5), 3)
    imgCanny = cv2.Canny(imgBlur, 50, 150)

    kernel = np.ones((3, 3), np.uint8)
    imgDilate = cv2.dilate(imgCanny, kernel, iterations=2)
    imgClosed = cv2.morphologyEx(imgDilate, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contours, _ = cv2.findContours(imgClosed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    print(f"Analysis result: A total of {len(contours)} objects were detected.")

    if len(contours) == 0:
        print("WARNING: No objects were found. The lighting may be very poor, or the image may not have loaded.")
        return

    count = 0
    saved_count = 0

    for cnt in contours:
        area = cv2.contourArea(cnt)

        if area > 800:
            x, y, w, h = cv2.boundingRect(cnt)

            # We place the money in a square and cut it out.
            imgCrop = img[y:y+h, x:x+w]

            # We automatically provide the file name.
            base_name = IMAGE_NAME.split('.')[0] # uzantıyı at
            filename = f"ref_{base_name}_{count}.jpg"

            cv2.imwrite(filename, imgCrop)
            print(f"-> {filename} kaydedildi. (Boyut: {int(area)})")
            saved_count += 1

            # Show on screen
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            count += 1

    print(f"\nA TOTAL OF {saved_count} REFERENCE IMAGES HAVE BEEN SAVED.")

    cv2_imshow(img)

if __name__ == "__main__":
    create_references()
```
